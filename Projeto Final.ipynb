Carregamos um clone do git do OpenIA Gym direto no repositório da maquina para que fosse feito vários testes com vários ambientes. Escolhemos o ambiente chamado "CartPole-v0".

Isso executará uma instância do CartPole-v0ambiente por 1000 etapas de tempo, renderizando o ambiente em cada etapa

```python
import gym
env = gym.make('CartPole-v0')
env.reset()
for _ in range(1000):
    env.render()
    env.step(env.action_space.sample()) # Realiza uma ação aleatória
env.close()
```

Nos exemplos acima, estivemos amostrando ações aleatórias do espaço de ação do ambiente. Mas o que realmente são essas ações? Cada ambiente vem com um action_spacee um observation_space. Esses atributos são do tipo Spacee descrevem o formato de ações e observações válidas:

```python
import gym
env = gym.make('CartPole-v0')
print(env.action_space)
#> Discrete(2)
print(env.observation_space)
#> Box(4,)
```

O Discrete espaço permite um intervalo fixo de números não negativos, portanto, neste caso, actions válidos são 0 ou 1. O Boxespaço representa uma n caixa     -dimensional, portanto, válido observations será uma matriz de 4 números. Também podemos verificar os Box limites de:

```python
print(env.observation_space.high)
#> array([ 2.4       ,         inf,  0.20943951,         inf])
print(env.observation_space.low)
#> array([-2.4       ,        -inf, -0.20943951,        -inf])
```

Essa introspecção pode ser útil para escrever código genérico que funcione para muitos ambientes diferentes. Boxe Discretesão os Spaces mais comuns . Você pode obter uma amostra de um Spaceou verificar se algo pertence a ele:

```python
from gym import spaces
space = spaces.Discrete(8) # Set with 8 elements {0, 1, 2, ..., 7}
x = space.sample()
assert space.contains(x)
assert space.n == 8
```
